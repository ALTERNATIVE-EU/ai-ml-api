## Import dataset


```python
def load_model_dataset_from_folder(folder_name):
    path = os.path.join(os.getcwd(), folder_name)
    folders = os.listdir(path)
    # data already as train and test
    dataset_models = {}
    for folder in folders:
        model_dict = {}
        print(folder)
        path_folder_models = os.path.join(path, folder)
        files = os.listdir(path_folder_models)
        for file in files:
            path_model = os.path.join(path_folder_models, file)
            if "dataset" not in file:
                try: model_dict['model'] = tf.keras.models.load_model(path_model)
                except: # if "saved_model" in file: model_dict['model'] = tf.saved_model.load(file)
                    print(file)            
            elif 'test' in file: model_dict['test'] = tf.data.Dataset.load(path_model)
            elif 'valid' in file: model_dict['valid'] = tf.data.Dataset.load(path_model)
            elif 'train' in file: model_dict['train'] = tf.data.Dataset.load(path_model)
        dataset_models[folder] = model_dict
    return dataset_models
```

```python
data = load_model_dataset_from_folder(MODEL_FOLDER)
```

```python
# test
n = 0
for name, model_and_data in data.items():
    print(name)
    try:
        pred = (model_and_data['model'].predict(model_and_data['test']) > 0.5).astype(int)
        print(sum(pred))
        y_test = [i[1] for i in list(model_and_data['test'].unbatch().as_numpy_iterator())]
        if n == 0:
            results_frame = pd.DataFrame(calculate_results(y_test, pred), index=[name])
        else:
            results = pd.DataFrame(calculate_results(y_test, pred), index=[name])
            results_frame = pd.concat([results_frame, results], axis=0)
        n += 1
    except: 
        print(name, "ERROR")
round(results_frame,3)            
```

```python
n = 0
for name, model_and_data in data.items():
    print(name)
    try:
        pred = (model_and_data['model'].predict(model_and_data['train']) > 0.5).astype(int)
        print(sum(pred))
        y_test = [i[1] for i in list(model_and_data['train'].unbatch().as_numpy_iterator())]
        if n == 0:
            results_frame = pd.DataFrame(calculate_results(y_test, pred), index=[name])
        else:
            results = pd.DataFrame(calculate_results(y_test, pred), index=[name])
            results_frame = pd.concat([results_frame, results], axis=0)
        n += 1
    except: 
            pass
round(results_frame,3)  
```

## Comparison different split

```python
index = ["split: 70-30", "split: 80-20", "split: 90-10"]

dict_results = {
    "XGB": [0.758, 0.747 , 0.742],
    "Multimodal": [0.743, 0.781, 0.808],
    "NLP Augmented": [0.759, 0.733, 0.815]}
```

```python
result_df = pd.DataFrame(dict_results, index=index).T
```

```python
import matplotlib.pyplot as plt
import numpy as np

species = result_df.index
penguin_means = result_df.copy()

x = np.arange(len(species))  # the label locations
width = 0.20 # the width of the bars
multiplier = 0

fig, ax = plt.subplots(layout='constrained', figsize=(25, 15))

for attribute, measurement in penguin_means.items():
    offset = width * multiplier
    rects = ax.bar(x + offset, measurement, width, label=attribute)
    ax.bar_label(rects, padding=2, fontsize=20)
    multiplier += 1

# Add some text for labels, title and custom x-axis tick labels, etc.
# ax.set_ylabel()
ax.set_title('Models behaviour: increasing train data', fontsize=40)
ax.set_xticks(x + width, species, fontsize=30)
ax.legend(loc='upper left', ncols=3)
ax.set_ylim(0, 1)
plt.ylabel('Balance Accuracy', fontsize=30)
# plt.xticks(rotation=45)
ax.tick_params(axis='y', labelsize=20)
plt.legend(loc="upper right", fontsize=20)
plt.savefig('high_resolution_plot.tif', dpi=300)
plt.show()
plt.close()
```
